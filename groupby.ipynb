```python
import pandas as pd
```

```python
df=pd.read_csv("/content/titanic.csv")
df
```

```python
avg_fare_class=df.groupby("Pclass")["Fare"].mean()
avg_fare_class
```

```python
gender_survived_most=df.groupby("Sex")["Survived"].sum()
gender_survived_most
```

```python
avg_fare=df.groupby(["Pclass","Sex"])["Fare"].mean()
avg_fare
```

```python
multiple_stats=df.groupby("Pclass").describe()
multiple_stats
```

```python
multiple_stats=df.groupby("Pclass").agg({
  # Average and median age per class
    "Fare": ["mean", "median","min","max","count"]  # Average and total fare per class
})


multiple_stats
```
